### Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df['normalization_function'] = scaler.fit_transform(df[['cost']])
```

### Standard scaler function

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
df['Standardization_function'] = std.fit_transform(df[['cost']])
df
```

### RobustScaler_function

```python
from sklearn.preprocessing import RobustScaler
```

```python
df['RobustScaler_function'] = RobustScaler().fit_transform(df[['cost']])
df
```

### Max abs

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
scaler = MaxAbsScaler()
df['scaler_abs_function'] = scaler.fit_transform(df[['cost']])
df
```
